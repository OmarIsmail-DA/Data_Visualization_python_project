```python
# Necessary imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

Define columns:

- make and model: Vehicle manufacturer and model.

- year: Year of manufacture.

- VClass: Vehicle class.

- drive: Drive type (e.g., All-Wheel Drive, Front-Wheel Drive).

- trans: Transmission type.

- fuelType: Type of fuel used.

- cylinders and displ: Number of cylinders and engine displacement.

- city, highway, comb: Fuel economy in city, highway, and combined, respectively.

- co2: CO2 emissions.
    
- feScore and ghgScore: Fuel economy and greenhouse gas scores.

```python
# Getting Data
df = pd.read_csv("fuel-econ.csv")
df
```

```python
df.info()
```

```python
# Choosing the Fig Size:
plt.figure(figsize = [14,6])
df['make'].value_counts().plot.bar()
```

```python
# Choosing the Figure Size
plt.figure(figsize=[14, 6])

# Grouping by 'make' and calculating the mean 'feScore'
avg_fescore_by_make = df.groupby('make')['feScore'].mean().sort_values()

# Plotting the Bar Chart
x = avg_fescore_by_make.index
y = avg_fescore_by_make.values

plt.bar(x, y)
plt.xlabel('Car Makes')
plt.ylabel('Average feScore')
plt.title('Average feScore by Car Makes')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.grid(axis='x', color='green', linestyle=':', linewidth=0.5)
plt.show()
```

```python
# Choosing the Figure Size
plt.figure(figsize=[14, 6])

# Grouping by 'make' and calculating the mean 'ghgScore'
avg_fescore_by_make = df.groupby('make')['ghgScore'].mean().sort_values()

# Plotting the Bar Chart
x = avg_fescore_by_make.index
y = avg_fescore_by_make.values

plt.bar(x, y)
plt.xlabel('Car Makes')
plt.ylabel('Average ghgScore')
plt.title('Average ghgScore Car Makes')
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.grid(axis='x', color='green', linestyle=':', linewidth=0.5)
plt.show()
```

```python
# Assuming df is your DataFrame
# Grouping by 'make' and calculating the mean 'highway' mileage, then selecting the top 10
top_10_makes = df.groupby('make')['highway'].mean().nlargest(10)

# Setting up the figure size
plt.figure(figsize=[14, 6])

# Data for plotting
x = top_10_makes.index
y = top_10_makes.values

# Creating the bar plot
plt.bar(x, y)

# Adding labels and title
plt.xlabel('Make')
plt.ylabel('Average Highway')
plt.title('Average Highway by Top 10 Car Makes')

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Adding a grid to the y-axis for clarity
plt.grid(axis='y', color='green', linestyle=':', linewidth=0.5)

# Display the plot
plt.show()
```

```python
# Assuming df is your DataFrame
# Counting the number of cars produced per year
make_counts_by_year = df.groupby('year')['make'].value_counts().unstack(fill_value=0)

# Summing up the counts of all makes for each year
total_make_counts_by_year = make_counts_by_year.sum(axis=1)

# Setting up the figure size
plt.figure(figsize=[14, 6])

# Plotting the line chart
plt.plot(total_make_counts_by_year.index, total_make_counts_by_year.values, marker='o')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of Car Makes')
plt.title('Total Car Makes Count by Year')
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Assuming df is your DataFrame
# Grouping by 'make' and calculating the average number of cylinders
avg_cylinders_by_make = df.groupby('make')['cylinders'].mean().sort_values()

# Setting up the figure size
plt.figure(figsize=[14, 6])

# Creating the bar plot
plt.bar(avg_cylinders_by_make.index, avg_cylinders_by_make.values)

# Adding labels and title
plt.xlabel('Car Makes')
plt.ylabel('Average Number of Cylinders')
plt.title('Average Number of Cylinders by Car Makes')
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```
